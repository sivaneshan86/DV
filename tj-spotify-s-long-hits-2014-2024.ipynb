```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS


text = " ".join(df["Name"])
wc = WordCloud(width = 300, height = 300).generate(text)

# Remove the axis and display the data as image
plt.axis("off")
plt.imshow(wc, interpolation = "bilinear")
plt.show() 
```

```python
#nltk librería de análisis de lenguaje
import nltk
#Este proceso puede hacerse antes de forma manual, descargar las stopwords de la librería nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words_sp = set(stopwords.words('spanish'))
stop_words_en = set(stopwords.words('english'))
stop_words = stop_words_sp | stop_words_en
```

```python
# Cleaning the stopwords
text = " ".join(df['Name'])
wc2 = WordCloud(width = 300, height = 300,
               stopwords = stop_words)
wc2.generate(text)

plt.axis("off")
plt.imshow(wc, interpolation = "bilinear")

# plt.show() 
```

```python
dict_word={}
for word in wc2.words_:
    #print(word)
    dict_word[word] = df[df["Name"].apply(lambda x: word in x)]
```

```python
keys_list = list(dict_word.keys())
filter_list=list(filter(lambda x: len(x)<3,keys_list))
for x in filter_list:
    keys_list.remove(x)
#keys_list
```

```python
dict_relation = {}
quiet_list = ['sleep','anxiety','relax','bird','peace','mind','sleep','dormir','white','help','rain','forest']
music_list = ['concert','violin','allegro','bird','concerto','wind','steel']

for name in keys_list:
    if 'mix' in name.lower() or 'version'  in name.lower() or 'remastered' in name.lower():
        dict_relation[name] = 'mix'
    for word in quiet_list:
        if word in name.lower():
            dict_relation[name] = 'quiet'
    for word in music_list:
        if word in name.lower():
            dict_relation[name] = 'concert'
```

```python
dict_relation
```

The mapping of info is superficial, so if we really want to investigate we should map every word on the map with a new category.

```python
df['new_col_aggr'] = ""
def extract_relation(x):
    for relation in dict_relation.keys():
        if relation in x:
            return(dict_relation[relation])
df['new_col_aggr'] = df['Name'].apply(lambda x : str(extract_relation(x)))
```

```python
value = []
for key in list(dict_relation.keys()):
    value += [dict_relation[key]]
value = set(value)
```

```python
# Histogram
for x in value:#dict_word.keys():
    fig, ax = plt.subplots()
    ax.hist(df[df['new_col_aggr'] == x]['Duration (Minutes)'])
    ax.set_title(x)
    plt.show()
```
